## SOV [81A]

```python
values_df = pd.read_csv("research/wals/wals-2020.1/cldf/values.csv", index_col="Language_ID")

# https://wals.info/languoid/lect/wals_code_dut
sov_values = values_df.loc[values_df.Parameter_ID == "81A"].Code_ID.map(codes_df)
sv_values = values_df.loc[values_df.Parameter_ID == "82A"].Code_ID.map(codes_df)
ov_values = values_df.loc[values_df.Parameter_ID == "83A"].Code_ID.map(codes_df)
oov_values = values_df.loc[values_df.Parameter_ID == "84A"].Code_ID.map(codes_df)
adp_values = values_df.loc[values_df.Parameter_ID == "85A"].Code_ID.map(codes_df)
gn_values = values_df.loc[values_df.Parameter_ID == "86A"].Code_ID.map(codes_df)
an_values = values_df.loc[values_df.Parameter_ID == "87A"].Code_ID.map(codes_df)
dn_values = values_df.loc[values_df.Parameter_ID == "88A"].Code_ID.map(codes_df)
nn_values = values_df.loc[values_df.Parameter_ID == "89A"].Code_ID.map(codes_df)
rn_values = values_df.loc[values_df.Parameter_ID == "90A"].Code_ID.map(codes_df)
qp_values = values_df.loc[values_df.Parameter_ID == "92A"].Code_ID.map(codes_df)
```

```python
rows = []

for lid, lang in lang_ids.items():
    sov = sov_values.get(lid, "Unknown")
    sv = sv_values.get(lid, "Unknown")
    ov = ov_values.get(lid, "Unknown")
    oov = oov_values.get(lid, "Unknown")
    adp = adp_values.get(lid, "Unknown")
    gn = gn_values.get(lid, "Unknown")
    an = an_values.get(lid, "Unknown")
    dn = dn_values.get(lid, "Unknown")
    nn = nn_values.get(lid, "Unknown")
    rn = rn_values.get(lid, "Unknown")
    qp = qp_values.get(lid, "Unknown")
    rows.append((lang, sov, sv, ov, oov, adp, gn, an, dn, nn, rn, qp))

order_df = pd.DataFrame(rows, columns=["language", "SOV", "SV", "OV", "OOV", "AdpNP", "GN", "AN", "DN", "NN", "RN", "QP"])
order_df.to_csv("research/udpos/word-order.csv")
order_df
```
